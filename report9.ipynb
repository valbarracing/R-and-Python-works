### 1.1 Aggregate Demand (AD)

Considering, on the one hand, the equation of the IS Curve:

$$ r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

- Where $ B_0 = C_o + I_o + G_o + X_o $ and $ B_1 = 1 - (b - m)(1 - t) $

And, on the other hand, the equation of the LM Curve

$$  r = -\frac{1}{j}\frac{Mo^s}{P} + \frac{k}{j}Y $$

And, eliminating $(r)$ and solving for $(P)$, we get:

$$  P = -\frac{h Mo^s}{-j B_o + (jB_1 + hk)Y} $$

Or, in function of $(Y)$:

$$  Y = \frac{jB_o}{jB_1 + hk} + (\frac{hMo^s}{jB_1 + hk})\frac{1}{P} $$


However, considering the money market equilibrium equation as follows:

$$ Mo^s - P = kY - jr $$

And solving $(r)$, we get the equation of the Aggregate Demand $(AD)$

$$  P = \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y $$


Therefore, the Aggregate Demand $(AD)$ is graphed:

```python
Y_size = 100

Ms = 1100  #nominal money supply
j = 10     #speculation money demand
k = 15     #transantion money demand

Co = 20   #consumo autonomo
Io = 10   #inversión autónoma
Go = 50   #gasto gobierno
Xo = 2    #exportacion
b = 0.4   #propension marginal a consumir
m = 0.2   #propension a importar
t = 0.9   #tasa impositiva
h = 1     #constante de decisión a invertir


B0 = Co + Io + Go + Xo
B1 = 1 - (b-m)*(1-t)

Y = np.arange(Y_size)
```

```python
 def P_AD(h, Ms, j, B0, B1, k, Y):
    P_AD = ((h*Ms + j*B0)/h) - (Y*(j*B1 + h*k)/h)
    return P_AD

P_AD = P_AD(h, Ms, j, B0, B1, k, Y)
P_AD
```

```python
y_max = np.max(P_AD)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Aggregate Demand", xlabel= r'Y', ylabel= r'P')
ax.plot(Y, P_AD, "k-", label = "AD")  
ax.legend()
plt.show()
```

### 1.2. Aggregate Supply (AS) in short-term:

The short term is a period in which the product $(Y)$ is located below or above its long-term level, called Potential Product or full employment $(\bar{Y})$.

Then, the Aggregate Supply $(AS)$ in the short run is represented:

$$ P = P^e + θ(Y - \bar{Y}) $$ 

- Where $(P)$ is the Price Level, $(P^e)$ the Expected Price and $\bar{Y}$ the Potential Product.

And the Aggregate Supply $(AS)$ is graphed as follows:

```python
Y_size = 100

Pe = 850   #precio esperado
θ = 23
_Y = 62   #producto potencial

Y = np.arange(Y_size)
```

```python
 def P_AS(Pe, _Y, Y, θ):
    P_AS = Pe + θ*(Y-_Y)
    return P_AS

P_AS = P_AS(Pe, _Y, Y, θ)
P_AS
```

```python
y_max = np.max(P_AS)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Aggregate Supply", xlabel= r'Y', ylabel= r'P')
ax.plot(Y, P_AS, "k-", label = "AS")  
ax.legend()
plt.show()
```

A decrease in production level will reduce the difference between production level $(Y)$ and potential product $(\bar{Y})$. Thus, the Price Level $(P)$ will fall.
 
In conclusion, a decrease in the money supply will generate a higher interest rate which will make production level and price level decrease .

In the context of the 2008 global economic crisis, US short-term interest rates reached levels close to zero percent. Therefore, because of this liquidity trap, the Federal Reserve Board (Fed) begins to implement a quantitative easing program: the purchase of public and private financial assets was promoted to increase the monetary base. In this connection, Mendonza's article aims to build a standard model of aggregate supply and demand, but considering tools of unconventional monetary policy. This, to include the quantitative easing innovations used by the Fed to counter the consequences of the 2008 economic crisis.  


The article explains and analyzes the three ways in which quantitative easing causes a fall in long-term interest rates and how the Fed, ceasing to be a lender of last resort, became a "market maker of last resort." The significance of this unconventional measure is even important in the current context or in any time of crisis. For example, the Fed again used quantitative easing as a strategy to mitigate the sharp contraction and deep uncertainty that the spread of COVID-19 caused. Hence, the importance of generating and studying a model that considers the short- and long-term bond market. 

On the one hand, the strength of the article is in using and demonstrating how old models and methods still work to analyze current macroeconomic problems. Mendoza points out that while the Fed's unconventional policy has been shaped in numerous jobs, they have done so using dynamic and stochastic general equilibrium models. However, the paper moves away from this obligatory instrumental of contemporary macroeconomists and adopts a more traditional theoretical option, as Keynesianism. 

On the other hand, three weaknesses are identified, not on the content of the article itself but on topics that could be included for a better understanding of quantitative easing. And, these ideas are also raised as future lines of research. 

First, it could be pointed out that this emerged as a strategy of emerging economies: quantitative easing was a monetary policy used by the Bank of Japan (BOJ) quantitative easing, before the US. Even nowadays, the BOJ has flooded commercial banks with excess liquidity to promote private lending and foster growth. 

Second, by arguing the validity and importance of traditional models, the article was able to elaborate more on the most contemporary models and the strengths and weaknesses of these to address the issue of unconventional monetary policy. With this comparison, one of the central arguments of the document would not only have had a strong support, but that more would have been known about what is said today about quantitative easing. 

Third, by focusing on the model built by Mendoza, perhaps it was possible to delve a little deeper into quantitative easing, its fundamentals, and its criticisms.
